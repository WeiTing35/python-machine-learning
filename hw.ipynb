```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
names = pd.read_csv('car.c45-names.txt',sep = '\t',encoding = 'utf-8',header=None)#read file
car_data = pd.read_csv('car.data.txt',sep = ",",encoding = 'utf-8')#read file
car_data.columns=["buying", "maint", "doors", "persons", "lug_boot", "safety", "class_values"]#add titles
```

```python
names #check file
```

```python
#check file
car_data
#buying,maint,doors,persons,lug_boot,safety,class_values
```

```python
#count the number of every pary
a=b=c=d=0 #initial count

#for loop to cal every row
for i in range(1727):
    if car_data.class_values[i] == 'unacc':
        a=a+1
    elif car_data.class_values[i] == 'acc':    
        b=b+1
    elif car_data.class_values[i] == 'good':    
        c=c+1    
    elif car_data.class_values[i] == 'vgood':    
        d=d+1

        
        
```

```python
#the count result
count=[a,b,c,d]
#total number
total=a+b+c+d 
#change to percent
count_p=[a*100/total , b*100/total,c*100/total,d*100/total]
#show
print(count_p)
```

```python
#plot size
plt.figure(figsize=(6,9))
#def labels
labels = [u'unacc',u'acc',u'good',u'vgood']
#percent for every parts
sizes = [count_p[0],count_p[1],count_p[2],count_p[3]]
#choose colors
colors = ['red','yellowgreen','lightskyblue','gold']

#explode the pie
explode = (0.05,0.05,0.05,0.05)

#define pit plot 
patches,l_text,p_text = plt.pie(sizes,explode=explode,labels=labels,colors=colors,
                                labeldistance = 1.1,autopct = '%3.1f%%',shadow = False,
                                startangle = 90,pctdistance = 0.6)


#set the size of the words
for t in l_text:
    t.set_size=(30)
for t in p_text:
    t.set_size=(20)
    

    

plt.axis('equal') #make the plot be a circle plot
plt.legend()
plt.show()#plot the result
```
